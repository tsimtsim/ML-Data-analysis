```python
import pandas as pd
df=pd.read_csv('titanic.csv')
df
```

```python
inputs = df.drop('Survived',axis='columns')
inputs
```

```python
from sklearn.preprocessing import LabelEncoder
le_Pclass = LabelEncoder()
le_Sex = LabelEncoder()
le_Age = LabelEncoder()
le_Fare = LabelEncoder()
```

```python
inputs['Pclass_n'] = le_Pclass.fit_transform(inputs['Pclass'])
inputs['Sex_n'] = le_Sex.fit_transform(inputs['Sex'])
inputs['Age_n'] = le_Age.fit_transform(inputs['Age'])
inputs['Fare_n'] = le_Fare.fit_transform(inputs['Fare'])
```

```python
inputs
```

```python
inputs2=inputs[['Pclass_n','Sex_n','Age_n','Fare_n']].copy()
```

```python
inputs2['Survived']=df['Survived']
```

```python
inputs2
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = inputs2.drop(['Survived'], axis='columns')
y = inputs2.Survived
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
len(X_test)
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
test=X_test.to_numpy()
test[1]
```

```python
model.predict([test[10]])
```

```python
y_test.to_numpy()[10]
```

Confusion Matrix------------------------------

```python

y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
